```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences'
             , 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism'
                 , 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education'
              , 'Agriculture','Business', 'Architecture']

fig = plt.figure(figsize=(15, 21))

for sp in range(0,18,3):
    stem_cats_i = int(sp/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[stem_cats_i]]
            , c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[stem_cats_i]]
            , c=cb_orange, label='Men', linewidth=3)
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[stem_cats_i])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    
    if sp == 0:
        ax.text(2005, 15, 'Men')
        ax.text(2002, 82, 'Women')
    elif sp == 15:
        ax.text(2005, 87, 'Men')
        ax.text(2001, 7, 'Women')
        
for sp in range(0,15,3):
    lib_arts_cats_i = int(sp/3)
    ax = fig.add_subplot(6,3,sp+2)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[lib_arts_cats_i]]
            , c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[lib_arts_cats_i]]
            , c=cb_orange, label='Men', linewidth=3)
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[lib_arts_cats_i])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    
    if sp == 0:
        ax.text(2005, 18, 'Men')
        ax.text(2002, 74, 'Women')
    elif sp == 12 :
        ax.text(2005, 54, 'Men')
        ax.text(2001, 40, 'Women')
        
for sp in range(0,18,3):
    other_cats_i = int(sp/3)
    ax = fig.add_subplot(6,3,sp+3)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[other_cats_i]]
            , c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[other_cats_i]]
            , c=cb_orange, label='Men', linewidth=3)
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[other_cats_i])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    
    if sp == 0:
        ax.text(2005, 7, 'Men')
        ax.text(2002, 90, 'Women')
    elif sp == 15:
        ax.text(2005, 63, 'Men')
        ax.text(2001, 30, 'Women')

plt.show()
```
